```python
sheets = ['CHN', 'LAO', 'KHM']
df = pd.DataFrame()
for sheet in sheets:
    temp_df = pd.read_excel('borders.xls', sheet_name=sheet)
    if sheet == 'CHN':
        temp_df['BorderWith'] = 'China'
    elif sheet == 'LAO':
        temp_df['BorderWith'] = 'Laos'
    else:
        temp_df['BorderWith'] = 'Cambodia'
    
    df = pd.concat([df, temp_df], ignore_index=True)
df
```
